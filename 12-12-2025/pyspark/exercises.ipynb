**Exercise 1:** Show the first 10 customers.

```python
df.show(10)

# df.limit(10).show()
```

**Exercise 2:** List all unique cities.

```python
df.select("city").distinct().show()
```

**Exercise 3:** Display only customer_id, name, and annual_spend columns.

```python
df.select(col("customer_id"), col("name"), col("annual_spend")).show()
```

**Exercise 4:** Filter all customers who spend more than 60000 annually.

```python
df.filter(col("annual_spend") > 6000).show()
```

**Exercise 5:** Show all customers from Delhi who are younger than 30.

```python
df.filter((col("city") == "Delhi") & (col("age") < 30)).show()
```

**Exercise 6:** Create a new column named:

*"spend_in_lakh" = annual_spend / 100000*.

```python
df.withColumn("spend_in_lakh", col("annual_spend")/100000).show()
```

**Exercise 7:** Create a new column "customer_type"

**Logic:**

1. spend > 70000 → Premium
2. else → Standard

```python
df.withColumn("customer_type", when(col("annual_spend") > 70000, "Premium").otherwise("Standard")).show()
```

**Exercise 8:** Show customers whose name starts with the letter A.

```python
df.filter(col("name").startswith("A")).show()
```

**Exercise 9:** Filter customers where category is either Clothing or Electronics.

```python
df.filter(col("categories").isin('Clothing', 'Electronics')).show()

# df.filter((df.categories  == "Electronics") | (df.categories  == "Clothing")).show()
```

**Exercise 10:** Convert the city name to uppercase.

```python
df.withColumn("city", upper(col("city"))).show()
```

**Exercise 11:** Remove the category column from the DataFrame.

```python
df.drop("categories").show()
```

**Exercise 12:** Sort customers by age in descending order.

```python
df.sort(col("age").desc()).show()
```

**Exercise 13:** Find customers who are younger than the average age of the entire dataset.

**Steps:**
1. Calculate average age.
2. Filter df where age < avg_age.

```python
avg_age_value = df.agg(avg(col("age"))).collect()[0][0]
df.filter(col("age") < avg_age_value).show()
```

**Exercise 14:** Display the top 5 highest spending customers.

```python
df.sort(col("annual_spend").desc()).limit(5).show()
```
