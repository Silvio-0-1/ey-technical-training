**Exercise 5:** Write only “Delivered” orders to JSON.

**Exercise 6:** Read JSON and filter:

* payment_mode = "Card"

```python
df_json = spark.read \
    .option("header","true") \
    .option("inferSchema","true") \
    .json("df_delivered.json")

df_json.filter((col("city") == "Mumbai") & (col("payment_mode") == "Card")).show()
```

**Exercise 7:** Add a column: *delivery_category*

Logic:
1. delivery_time_minutes > 45 → "Late"
2. else → "OnTime"


Write output to JSON.

```python
df = df.withColumn("delivery_category", when(col("delivery_time_minutes") > 45, "Late").otherwise("OnTime"))
df.show()

df.write.mode("overwrite").json("df_added_column.json")
```

**Exercise 8**: Force JSON output to a single partition and observe number of files created.

```python
df_json = df_json = spark.read \
    .option("header","true") \
    .option("inferSchema","true") \
    .json("df_added_column.json")

df_json.repartition(1).write.mode("overwrite").json("df_single_partition.json")
```

**Exercise 15:** Which cuisine generates the highest order_amount overall?

```python
df.groupBy("cuisine").agg(sum("order_amount").alias("Total")).orderBy("Total", ascending = False).show()
```

**Exercise 16:** Which city has the highest number of orders?

```python
df.groupBy("city").agg(count("order_id").alias("Number of Orders")).orderBy("Number of Orders", ascending = False).show()
```

**Exercise 17:** Which payment mode is most frequently used?

```python
df.groupBy("payment_mode").agg(count("payment_mode").alias("Number of Payments")).orderBy("Number of Payments", ascending = False).show()
```
