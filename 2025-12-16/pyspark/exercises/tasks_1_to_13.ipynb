**Task 1:** Read the CSV file using sparkContext.textFile and display the first 5 records.

```python
with open('call_records.csv','w') as f:
    f.write(data)

rdd = spark.sparkContext.textFile('call_records.csv')
```

```python
# first 5 rows
rdd.take(5)
```

**Task 2:** Remove the header row and create a clean RDD containing only data rows.

```python
header = rdd.first()
data_rdd = rdd.filter(lambda x: x != header)

data_rdd.collect()
```

**Task 3:** Split each row into individual fields using a delimiter.

```python
split_rdd = data_rdd.map(lambda x: x.split(','))
split_rdd.collect()
```

**Task 4:** Calculate the total call cost per city.

```python
city_cost = split_rdd.map(lambda x: (x[3], float(x[6])))

# city_cost.collect() # displays "city", "cost" columns

total_cost_per_city = city_cost_rdd.reduceByKey(lambda a, b: a + b) # adds all "cost" belonging to same city

total_cost_per_city.collect()
```

**Task 5:** Identify the city with the highest total call cost.

```python
highest_cost = total_cost_per_city.reduce(lambda x, y: x if x[1]> y[1] else y) # compares "cost" column of two rows and returns the highest
highest_cost
```

**Task 7:** Count the number of calls per city.

```python
calls_1 = split_rdd.map(lambda x: (x[3], 1))

# calls_1.collect() # displays as "city", 1

calls_per_city = calls_1.reduceByKey(lambda a, b: a + b) # adds all 1 belonging to same city

calls_per_city.collect()
```

**Task 8:** Calculate the average call cost per city using RDD transformations.

```python
avg_cost_per_city = total_cost_per_city.join(calls_per_city).map(lambda x: (x[0], x[1][0] / x[1][1]))
avg_cost_per_city.collect()
```
