**Exercise 7:** Add a column: *delivery_category*

**Exercise 18:** Why is Parquet the preferred format for analytics platforms like Databricks and BigQuery?

**Ans:** Parquet is preferred for analytics platforms like **Databricks** and **BigQuery** because:

1.  **Columnar Storage** – It stores data by columns, which is ideal for analytical queries that often scan only a few columns instead of entire rows.
2.  **Efficient Compression & Encoding** – Reduces storage size and speeds up I/O operations.
3.  **Schema Support** – Handles complex data types and nested structures, making it flexible for big data.
4.  **Predicate Pushdown** – Filters data at the storage level, minimizing data read and improving query performance.
5.  **Optimized for Parallel Processing** – Works well with distributed systems like Spark and BigQuery for large-scale analytics.
