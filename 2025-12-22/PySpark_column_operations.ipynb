```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("MyApp") \
    .getOrCreate()
```

```python
data=[
    ("U001", "Delhi", 450),
    ("U002", "Mumbai", 620),
    ("U003", "Bangalore", 300),
    ("U004", "Delhi", 700)
]
```

```python
columns=["user_id", "city", "amt"]
```

```python
df=spark.createDataFrame(data, columns)
df.show()
```

```python
from pyspark.sql.functions import col
df=df.withColumn("amt_with_tax", col("amt")*1.18)
```

```python
df.show()
```

```python
df=df.replace("Delhi", 'New Delhi', subset=['city'])
```

```python
df.show()
```

```python
from pyspark.sql.functions import when
```

```python
df=df.withColumn(
    "amt_cat",
    when(col("amt") >= 500, "High").otherwise("Low")
)
```

```python
df.show()
```
