**Exercise 15:** Which city generates the highest total trip_fare?

**Exercise 16:** Which vehicle_type has the highest average fare?

```python
df.groupBy("vehicle_type").agg(round(avg("trip_fare"), 2).alias("Average Fare")).orderBy("Average Fare", ascending = False).show()
```

**Exercise 17:** Which driver has completed the most trips?

**Exercise 18:** Why is Parquet preferred for analytics dashboards and aggregations?

Parquet is preferred for **analytics dashboards and aggregations** because:

1.  **Columnar Format** – Aggregations (like SUM, AVG, COUNT) typically operate on specific columns. Parquet reads only those columns instead of entire rows, reducing I/O.
2.  **Compression & Encoding** – Smaller size means faster data transfer and lower storage costs.
3.  **Predicate Pushdown** – Filters are applied at the storage level, so dashboards load faster by scanning less data.
4.  **Optimized for Parallel Processing** – Works efficiently with distributed engines (Spark, Presto, BigQuery), which dashboards often rely on.
5.  **Schema Evolution** – Handles complex and evolving data structures, useful for dynamic dashboards.
